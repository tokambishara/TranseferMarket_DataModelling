```python
import pandas as pd
df_games=pd.read_csv(r'C:\Users\pc\Desktop\transfer\games.csv')
df_games['games_sk']=range(1, len(df_games) + 1)
df_games = df_games[[df_games.columns[-1]] + list(df_games.columns[:-1])]
df_games.head()
```

```python
df_clubs=pd.read_csv(r'C:\Users\pc\Desktop\transfer\clubs.csv')
df_clubs['clubs_sk']=range(1, len(df_clubs) + 1)
df_clubs = df_clubs[[df_clubs.columns[-1]] + list(df_clubs.columns[:-1])]
df_clubs.head()
```

```python
df_player=pd.read_csv(r'C:\Users\pc\Desktop\transfer\players.csv')
df_player['player_sk']=range(1, len(df_player) + 1)
df_player = df_player[[df_player.columns[-1]] + list(df_player.columns[:-1])]
df_player.head()
```

```python
df_competitions=pd.read_csv(r'C:\Users\pc\Desktop\transfer\competitions.csv')
df_competitions['competitions_sk']=range(1, len(df_competitions) + 1)
df_competitions = df_competitions[[df_competitions.columns[-1]] + list(df_competitions.columns[:-1])]
df_competitions.head()
```

```python
df_games = pd.merge(df_games, df_competitions[['competition_id', 'name']], on='competition_id').drop(columns=['competition_id'])
df_games.rename(columns={'name': 'competition_name'}, inplace=True)
df_games.drop(columns=['home_club_id', 'away_club_id'], inplace=True)
df_games.columns
```

```python
df_clubs.rename(columns={'name': 'club_name'}, inplace=True)
df_clubs = pd.merge(df_clubs,df_competitions[['competition_id','name']], left_on='domestic_competition_id',right_on='competition_id').drop(columns=['competition_id'])
df_clubs.drop(columns=['domestic_competition_id'], inplace=True)
df_clubs.rename(columns={'name': 'domestic_competition_name'}, inplace=True)
df_clubs.columns



```

```python
df_player.rename(columns={'name': 'player_name'}, inplace=True)
df_player = pd.merge(df_player,df_competitions[['competition_id','name']], left_on='current_club_domestic_competition_id',right_on='competition_id').drop(columns=['competition_id'])
df_player.drop(columns=['current_club_domestic_competition_id','current_club_id'], inplace=True)
df_player.rename(columns={'name': 'current_club_domestic_competition_name'}, inplace=True)
df_player.columns
```

```python
df_competitions.drop(columns=['country_id'], inplace=True)
df_competitions.rename(columns={'name': 'competition_name'}, inplace=True)
df_competitions.columns
```

```python
fact_apperance=pd.read_csv(r'C:\Users\pc\Desktop\transfer\appearances.csv')
fact_apperance.head()
```

```python
fact_apperance = pd.merge(fact_apperance,df_games[['game_id', 'games_sk']],on='game_id').drop(columns=['game_id'])
fact_apperance = pd.merge(fact_apperance,df_clubs[['club_id', 'clubs_sk']],left_on='player_current_club_id',right_on='club_id').drop(columns=['club_id','player_current_club_id'])  
fact_apperance = pd.merge(fact_apperance,df_competitions[['competition_id', 'competitions_sk']],on='competition_id',how='inner').drop(columns=['competition_id'])  
fact_apperance = pd.merge(fact_apperance,df_player[['player_id', 'player_sk']],on='player_id',how='inner').drop(columns=['player_id'])  
fact_apperance['Primary_Key']=range(1, len(fact_apperance) + 1)
fact_apperance=fact_apperance[['Primary_Key','appearance_id','player_sk','games_sk','clubs_sk','competitions_sk','date','yellow_cards','red_cards','goals','assists','minutes_played']]
fact_apperance.columns
```

```python
fact_apperance.to_csv('fact_apperance.csv')
df_player.to_csv('Dim_Players.csv')
df_clubs.to_csv('Dim_clubs.csv')
df_competitions.to_csv('Dim_competitions.csv')
df_games.to_csv('Dim_games.csv')
```
